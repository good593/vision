# 이미지
- https://blog.naver.com/samsjang/220499281999

```python
def showImage():
  imgfile = './data/img/wonman_02.png'
  img = cv2.imread(imgfile, cv2.IMREAD_COLOR)
  cv2.imshow('model', img)

  k = cv2.waitKey(0) & 0xFF
  if k == 27:
    cv2.destroyAllWindows()
  elif k == ord('c'):
    cv2.imwrite('./data/save/wonman_02.png', img) # 이미지 저장
    cv2.destroyAllWindows()

# showImage()
```

### img = cv2.imread(imgfile, cv2.IMREAD_COLOR)
> cv2의 imread 함수는 이미지 파일을 읽기 위한 객체를 리턴합니다. 이 함수의 첫번째 인자는 읽고자 하는 파일의 경로, 두번째 인자는 이미지 파일을 읽는 방식을 나타내는 플래그입니다.  
  
이미지 읽기 플래그는 아래와 같이 3가지가 있습니다.
- `cv2.IMREAD_COLOR`: 컬러 이미지로 로드함. 이미지의 투명한 부분은 모두 무시됨. 디폴트 플래그. 정수값은 1
- `cv2.IMREAD_GRAYSCALE`: 흑백 이미지로 로드함. 정수값은 0
- `cv2.IMREAD_UNCHANGED`: 알파채널(Alpha-channel)을 포함하여 이미지 그대로 로드함. 정수값은 -1
### cv2.imshow('model', img)
> cv2.imread()에 의해 반환된 이미지 객체 img를 화면에 나타내기 위한 함수입니다. 이 함수의 첫 번째 인자는 윈도우 타이틀이며, 두번째 인자는 화면에 표시할 이미지 객체입니다.
### cv2.waitKey(0) & cv2.destroyAllWindows()
> 화면에 이미지를 표시한 후 사용자가 키보드를 누를 때까지 대기하라는 코드입니다.
- `cv2.waitKey()`: 지정된 시간동안 키보드 입력을 기다리는 함수입니다. 인자는 키보드 입력을 기다리는 시간으로 1/1000 초(ms)입니다. 
  - cv2.waitKey(0): 키보드 입력이 있을 때까지 무제한으로 기다리라는 의미
  - ESC키를 누르면, cv2.waitKey()함수는 27을 리턴하게 됩니다.  
- `cv2.destroyAllWindows()`: 생성한 모든 윈도우를 제거합니다.
### Matplotlib을 이용해서 화면에 이미지 표시하기

```python
def showImageWithPlt():
  imgfile = './data/person_01.png'
  img = cv2.imread(imgfile, cv2.IMREAD_GRAYSCALE)
  
  plt.imshow(img, cmap='gray', interpolation='bicubic')
  plt.show()

showImageWithPlt()
```

# 비디오

```python
video_file = './data/Roller Coaster.mp4' # 동영상 파일 경로

cap = cv2.VideoCapture(video_file)  # 동영상 캡쳐 생성
if not cap.isOpened(): # 동영상 캡쳐 초기화 실패
  print('can not open video.')

while True:
  ret, img = cap.read() # 프레임 읽기
  if not ret: # 다음 프레임 없음
    break 

  cv2.imshow(video_file, img) # 화면에 표시
  cv2.waitKey(23) # 25ms 지연

cap.release() # 캡쳐 자원 반납
cv2.destroyAllWindows()
```
